```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:\\Users\\AkankshaAmarendra\\PycharmProjects\\pythonProject1\\homeprices.csv")
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area, df.price,color='red',marker='x')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df['price'])
```

```python
predicted_price = reg.predict(np.array([[3100]]))  # 2D array with a single value
print(predicted_price)
```
