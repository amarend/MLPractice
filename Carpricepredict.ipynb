```python
import pandas as pd
df = pd.read_csv('carprices.csv')
df
```

```python
dummis = pd.get_dummies(df['Car Model'])
merged = pd.concat([df, dummis.astype(int)],axis='columns')
merged
```

```python
final_removed = merged.drop(columns=['Car Model', 'Audi A5'])
final_removed
```

```python
X = final_removed.drop(columns=['Sell Price($)'])
X
```

```python
y = final_removed['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
model.score(X,y)
```
